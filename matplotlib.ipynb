```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batter = pd.read_csv('Datasets/batter.csv')
batter.head(5)
```

```python
# Plotting random top 100, 25 samples of scatter plot

sample_df = batter.head(100).sample(25)
sample_df
```

```python
plt.scatter(x=sample_df['avg'], y=sample_df['strike_rate'])
```

```python
plt.scatter(x=sample_df['avg'], y=sample_df['strike_rate'])
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Batter's INFO")
```

```python
a = [1,3,5,7]
b = [2,4,6,8]

plt.figure(figsize = (15,7))

plt.scatter(a,b)
plt.text(x=1.1, y = 2.1, s="(1,2)")
plt.text(x=3.1, y = 4.1, s="(3,4)")
plt.text(x=5.1, y = 6.1, s="(5,6)")
plt.text(x=7.1, y = 8.1, s="(7,8)")
plt.show()
```

```python
# Annotating the batter data in scatter plot
plt.figure(figsize=(18,12))
plt.scatter(x=sample_df['avg'], y=sample_df['strike_rate'], s=sample_df['runs']/10)

plt.axhline(130,color="Green")
plt.axvline(30, color='green')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] + 0.1, sample_df['strike_rate'].values[i] + 0.1, sample_df['batter'].values[i],
             fontdict={'color':'red',
                       'size': 13})

plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Batter's INFO")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(sample_df['avg'], sample_df['strike_rate'])
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_title("Batter's Info")
fig.show()
```
