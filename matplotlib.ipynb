```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batter = pd.read_csv('Datasets/batter.csv')
batter.head(5)
```

```python
# Plotting random top 100, 25 samples of scatter plot

sample_df = batter.head(100).sample(25)
sample_df
```

```python
plt.scatter(x=sample_df['avg'], y=sample_df['strike_rate'])
```

```python
plt.scatter(x=sample_df['avg'], y=sample_df['strike_rate'])
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Batter's INFO")
```

```python
a = [1,3,5,7]
b = [2,4,6,8]

plt.figure(figsize = (15,7))

plt.scatter(a,b)
plt.text(x=1.1, y = 2.1, s="(1,2)")
plt.text(x=3.1, y = 4.1, s="(3,4)")
plt.text(x=5.1, y = 6.1, s="(5,6)")
plt.text(x=7.1, y = 8.1, s="(7,8)")
plt.show()
```

```python
sample_df.head()
```

```python
# Annotating the batter data in scatter plot
plt.figure(figsize=(18,12))
plt.scatter(x=sample_df['avg'], y=sample_df['strike_rate'], s=sample_df['runs']/10)

plt.axhline(130,color="Green")
plt.axvline(30, color='green')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] + 0.1, sample_df['strike_rate'].values[i] + 0.1, sample_df['batter'].values[i],
             fontdict={'color':'red',
                       'size': 13})

plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Batter's INFO")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(sample_df['avg'], sample_df['strike_rate'])
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_title("Batter's Info")
fig.show()
```

```python
season = pd.read_csv('Datasets/batsman_season_record.csv')
season
```

```python
plt.bar(season['batsman'],season['2015'],color='black')
plt.xticks(rotation=40)
plt.show()
```

```python
plt.barh(season['batsman'],season['2016'],color='pink')
```

```python
season
```

```python
# Multiple Bar Chart

plt.bar(np.arange(5)-0.2, season['2015'],width=0.2,label='2015')
plt.bar(np.arange(5), season['2016'],width=0.2,label='2016')
plt.bar(np.arange(5)+0.2, season['2017'],width=0.2,label='2017')

plt.xticks(np.arange(5), season['batsman'],size=10,rotation=30)
plt.xlabel("Players")
plt.ylabel("Run Scored")
plt.legend()
```

```python
season['2015']+season['2016']+season['2017']
```

```python
# Stacked Bar Chart

plt.bar(season['batsman'], season['2015'],label='2015')
plt.bar(season['batsman'], season['2016'],label='2016',bottom=season['2015'])
plt.bar(season['batsman'], season['2017'],label='2017',bottom=season['2015']+ season['2016'])


plt.grid(axis='y')
plt.legend()
```
